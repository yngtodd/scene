### Preparing External Data: Cornell Movie Dialogs

```python
import os
import pandas as pd
```

```python
path = '/Users/yngtodd/data/cornell_movies'
movie_lines = os.path.join(path, 'movie_lines.txt')
```

```python
def get_id2line(lines_path):
    with open(lines_path, 'rb') as f:
         lines = [l.decode('utf8', 'ignore') for l in f.readlines()]

    id2line = {}
    for line in lines:
        _line = line.split(' +++$+++ ')
        if len(_line) == 5:
            id2line[_line[0]] = _line[4]
    return id2line
```

```python
def get_conversations(lines_path):
    with open(lines_path, 'rb') as f:
         conv_lines = [l.decode('utf8', 'ignore') for l in f.readlines()]
    #conv_lines = open('movie_conversations.txt').read().split('\n')
    convs = [ ]
    for line in conv_lines[:-1]:
        _line = line.split(' +++$+++ ')[-1][:-1]#.replace("'","")#.replace(" ","")
        convs.append(_line.split(','))
    return convs
```

```python
def extract_conversations(convs,id2line,path='conversations'):
    idx = 0
    for conv in convs:
        f_conv = open(path + str(idx)+'.txt', 'w')
        for line_id in conv:
            f_conv.write(id2line[line_id])
            f_conv.write('\n')
        f_conv.close()
        idx += 1
```

```python
def gather_dataset(convs, id2line):
    questions = []; answers = []

    for conv in convs:
        if len(conv) %2 != 0:
            conv = conv[:-1]
        for i in range(len(conv)):
            if i%2 == 0:
                questions.append(id2line[conv[i]])
            else:
                answers.append(id2line[conv[i]])

    return questions, answers
```
